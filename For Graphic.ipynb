```python
import pandas as pd
import numpy as np
import matplotlib as mlt

df = pd.read_csv('clean.csv')

df.head(3)
```

```python
df['VALUES'] = df.groupby(['CLEANED CRIME'])['YEAR'].transform('count')

```

```python
def custom_round(x, base=5):
    return int(base * round(float(x)/base))

df['PERIOD'] = df['YEAR'].apply(lambda x: custom_round(x, base=5))
df['CRIME SLUG'] = df['CLEANED CRIME'].str.replace(' ','-')
df['CRIME SLUG'] = df['CRIME SLUG'].str.replace('---','-')
df['CRIME SLUG'] = df['CRIME SLUG'].str.lower()

df.head(10)
```

```python
dfmod = df[df['VALUES'] > 49]

#dfmod.to_csv('what-is-this.csv')
```

```python
newpivot = dfmod.pivot_table(index=['PERIOD'],values='VALUES',columns=['CLEANED CRIME'],aggfunc='count')

newpivot.to_csv('output_for_graphic.csv')
```

```python
methods = df.pivot_table(index=['PERIOD'],values='VALUES',columns=['METHOD CLEAN'],aggfunc='count')

methods.to_csv('methods_for_graphic.csv')
```
